```python
import pandas as pd
import geocoder
```

```python
%%time
sheets = pd.read_excel('in/Planilhas_Enem_2015_download.xlsx', sheetname=None)
```

```python
# Comum a todas as disciplinas.
columns_com = ['Codigo','Nome','UF','Municipio','Administracao','Local',
        'QtdAlunosCenso','PorteEscola','QtdParticipantes','TaxaParticipacao',
        'QtdParticipantesEspeciais','IndPermanenciaEscola','IndSocioEcon',
        'FaixaIndFormDocente','IndFormDocente','TaxaAprovacao','TaxaReprovacao',
        'TaxaAbandono']

# Específico por tipo de disciplina.
columns_esp = ['Media30','Media','PercNivel1','PercNivel2',
        'PercNivel3','PercNivel4','PercNivel5' ]

columns = columns_com + columns_esp
```

```python
sheets['RD'] = sheets['Redação']
sheets.pop('Redação');
```

```python
escolas = None
already = False
for k,v in sheets.items():
    v.columns = columns
    v.drop(0, inplace=True)
    v.index = v.Codigo
    if not already:
        escolas = v[columns_com].copy()
        already = True
    v.drop(columns_com, axis=1, inplace=True)
```

```python
pn = pd.Panel.from_dict(sheets)
pn
```

```python
pn = pn.transpose('items', 'minor_axis', 'major_axis')
pn
```

```python
idx_rio = escolas.query('UF == "RJ" and Municipio == "Rio de Janeiro"').index
```

```python
idx_rio
```

```python
escolas['MediaTotal'] = pn.major_xs('Media').sum(axis=1)
```

```python
rio = escolas.ix[idx_rio]
```
